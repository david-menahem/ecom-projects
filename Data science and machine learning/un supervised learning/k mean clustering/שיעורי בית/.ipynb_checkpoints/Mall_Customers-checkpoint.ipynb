```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers_df_original = pd.read_csv('Mall_Customers.csv')
customers_df = customers_df_original.copy()
customers_df.head()
```

```python
customers_df = customers_df.drop('CustomerID', axis=1)
customers_df
```

```python
customers_df.isna().sum()
```

```python
customers_df.duplicated().unique()
```

```python
unique_values = customers_df.copy()
for col in customers_df.columns:
    unique_values[col] = pd.Series(customers_df[col].unique())
unique_values
```

```python
customers_df = pd.get_dummies(customers_df,columns=['Genre'],drop_first=True)
customers_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
scaled_customer_df = scaler.fit_transform(customers_df)
scaled_customer_df
```

```python
kmeans_model = KMeans(n_clusters=4)
y_pred = kmeans_model.fit_predict(scaled_customer_df)
len(y_pred)
```

```python
kmeans_model.cluster_centers_
```

```python
y_pred
```

```python
sns.scatterplot(y=np.linspace(1,201,200),x=y_pred)
plt.ylabel('CustomerID')
plt.xlabel('Cluster')
plt.title('Cluster scaterrplot by customer ID')
plt.show()
```

```python
ssd = []
for k in range(2,11):
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(scaled_customer_df)
    ssd.append(kmeans_model.inertia_)
ssd
```

```python
plt.plot(np.linspace(2,11,9),ssd,marker='o')
plt.show()
```

# The lowest k is 9 but the ssd decreaed very litlle by adding clusters above 6 and we are considerring customers groups
# since I dont't want to many groups or overfitting I will choose 6

```python
optimal_k = 6
```

```python
final_kmeans_model = KMeans(n_clusters=optimal_k)
scaler = StandardScaler()
scaler.fit(customers_df)
scaled_customers_df = scaler.transform(customers_df)
final_kmeans_model.fit(scaled_customers_df)
y_pred = final_kmeans_model.predict(scaled_customers_df)
customers_df_with_cluster = customers_df.copy()
customers_df_with_cluster['Cluster'] = y_pred
customers_df_with_cluster
```

```python
# not really neccesary is this case just for fun
def dummies(new_df,old_df):
    new_data_with_dummy_columns = pd.get_dummies(new_df,drop_first=True)
    missing_dummy_columns = old_df.columns.difference(new_data_with_dummy_columns.columns)
    for col in missing_dummy_columns:
        new_data_with_dummy_columns[col] = False
    new_data_with_dummy_columns = new_data_with_dummy_columns[old_df.columns]
    return new_data_with_dummy_columns
dummie = dummies    
```

```python
from joblib import dump, load
dump(scaler,'scaler.joblib')
dump(dummie,'dummies.joblib')
dump(final_kmeans_model,'kmean_customers.joblib')
```

```python
loaded_scaler = load('scaler.joblib')
loaded_dummies = load('dummies.joblib')
loaded_model = load('kmean_customers.joblib')
```

```python
new_data = {'CustomerID':[201,202,203],'Genre':['Male','Female','Male'],'Age':[45,28,33],'Annual Income (k$)':[20,22,19],'Spending Score (1-100)':[45,92,20]}
new_data_df = pd.DataFrame(new_data)
new_data_df.drop('CustomerID',axis=1)
new_data_df_dummies = loaded_dummies(new_data_df,customers_df)
scaled_new_data_df = loaded_scaler.transform(new_data_df_dummies)
y_pred = loaded_model.predict(scaled_new_data_df)
y_pred
```
