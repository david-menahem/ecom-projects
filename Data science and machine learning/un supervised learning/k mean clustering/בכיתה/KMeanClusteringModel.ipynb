```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../bank-full.csv')
df
```

```python
df['pdays'].unique()
# df['education'].unique()
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.histplot(data=df, x='age', bins=40, hue='loan')
plt.show()
```

```python
data_df = df[df['pdays'] != 999]
plt.figure(figsize=(12,6), dpi=200)
sns.histplot(data=data_df, x='pdays', bins=40)
plt.show()
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.histplot(data=df, x='duration', bins=100, hue='contact')
plt.show()
```

```python
order_job_df = df['job'].value_counts().index
plt.figure(figsize=(12,6), dpi=200)
sns.countplot(data=df, x='job', order=order_job_df)
plt.xticks(rotation=90)
plt.show()
```

```python
order_education_df = df['education'].value_counts().index
plt.figure(figsize=(12,6), dpi=200)
sns.countplot(data=df, x='education', order=order_education_df)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = pd.get_dummies(df)
scaler = StandardScaler()

scaled_x = scaler.fit_transform(X)

model = KMeans(n_clusters=3)

cluster_labels = model.fit_predict(scaled_x)
cluster_labels
```

```python
X = pd.get_dummies(df)
scaler = StandardScaler()

scaled_x = scaler.fit_transform(X)

ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_x)
    
    ssd.append(model.inertia_)
    
ssd
```

```python
plt.plot(range(2,10), ssd, marker='o', linestyle='dashed')
plt.show()
```
