```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tree_blobs = pd.read_csv('tree_blobs_with_outlier.csv')
tree_blobs
```

```python
tree_blobs = tree_blobs.drop('Unnamed: 0',axis=1)
```

```python
sns.scatterplot(data=tree_blobs,x='X1',y='X2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN()
labels = model.fit_predict(tree_blobs)
labels
```

```python
sns.scatterplot(data=tree_blobs,x='X1',y='X2',hue=labels)
plt.show()
```

```python
all_count_outliers = []
for eps in np.linspace(0.001,10,1000):
    model = DBSCAN(eps=eps)
    labels = model.fit_predict(tree_blobs)
    count_outliers = np.sum(labels==-1)
    all_count_outliers.append(count_outliers)
```

```python
plt.plot(np.linspace(0.001,10,1000),all_count_outliers)
plt.hlines(y=10,xmin=0,xmax=1,color='red')
plt.xlabel('Eps')
plt.ylabel('Outliers')
plt.xlim(0,1)
plt.ylim(0,20)
plt.title("No' of outliers for each epsilon value")
plt.show()
```

```python
all_count_outliers = []
for min_samples in range(1,10):
    model = DBSCAN(min_samples=min_samples)
    labels = model.fit_predict(tree_blobs)
    count_outliers = np.sum(labels==-1)
    all_count_outliers.append(count_outliers)
```

```python
# The goals is to find a model with 10 outliers
```

```python
plt.plot(np.linspace(1,10,9),all_count_outliers)
plt.xlabel('Min samples')
plt.ylabel('Outliers')
plt.title("No' of outliers for each min sample value")
plt.show()
```

```python
model = DBSCAN(eps=0.6,min_samples=2)
labels = model.fit_predict(tree_blobs)
sns.scatterplot(data=tree_blobs,x='X1',y='X2',hue=labels)
plt.show()
outliers = np.sum(labels == -1)
outliers
```

```python
# when using a combination of the optimal min samples and the optimal epsilon, the number of outliers reduced to 4
```

```python
model = DBSCAN(eps=0.15,min_samples=3)
labels = model.fit_predict(tree_blobs)
sns.scatterplot(data=tree_blobs,x='X1',y='X2',hue=labels,palette='viridis')
plt.show()
```

```python
# I reduced both parameters and found a plot with multiple clusters and no outliers
```
