```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs
```

```python
sns.scatterplot(data=blobs,x='X1',y='X2')
plt.show()
```

```python
moons = pd.read_csv('cluster_moons.csv')
circles = pd.read_csv('cluster_circles.csv')
circles
```

```python
sns.scatterplot(data=moons,x='X1',y='X2')
plt.show()
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
plt.show()
```

```python
def display_scatterplot(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
    plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
display_scatterplot(model,blobs)
```

```python
model = KMeans(n_clusters=2)
display_scatterplot(model,moons)
```

```python
display_scatterplot(model,circles)
```

```python
from sklearn.cluster import KMeans,DBSCAN
model = DBSCAN(eps=2)
display_scatterplot(model,blobs)
model = DBSCAN(eps=0.15)
display_scatterplot(model,moons)
display_scatterplot(model,circles)
```

```python
two_blobs_outliers = pd.read_csv('cluster_two_blobs_outliers.csv')
sns.scatterplot(data=two_blobs_outliers,x='X1',y='X2')
plt.show()
```

```python
model = DBSCAN()
display_scatterplot(model,two_blobs_outliers)
```

```python
model = DBSCAN(eps=0.001)
display_scatterplot(model,two_blobs_outliers)
```

```python
model = DBSCAN(eps=10)
display_scatterplot(model,two_blobs_outliers)
```

```python
model = DBSCAN(min_samples=1)
display_scatterplot(model,two_blobs_outliers)
```

```python
model = DBSCAN(min_samples=10)
display_scatterplot(model,two_blobs_outliers)
```

```python
model = DBSCAN(eps=1)
display_scatterplot(model,two_blobs_outliers)
```

```python
eps = np.linspace(0.001,10,1000)
percentages = []
outliers = []
for e in eps: 
    model = DBSCAN(eps=e)
    model.fit(two_blobs_outliers)
    labels = model.labels_
    num_outliers = np.sum(labels==-1)
    outliers.append(num_outliers)
    outliers_percentage = num_outliers*100/len(labels)
    percentages.append(outliers_percentage)
```

```python
plt.plot(eps,outliers)
plt.xlim(0,1)
plt.ylim(0,10)
plt.hlines(y=3,xmin=0,xmax=1,colors='red')
plt.show()
```

```python
mins = range(1,1000)
percentages = []
outliers = []
for min in mins: 
    model = DBSCAN(min_samples=min)
    model.fit(two_blobs_outliers)
    labels = model.labels_
    num_outliers = np.sum(labels==-1)
    outliers.append(num_outliers)
    outliers_percentage = num_outliers*100/len(labels)
    percentages.append(outliers_percentage)
```

```python
plt.plot(mins,outliers)
plt.ylim(0,10)
plt.xlim(0,5)
plt.show()
```
