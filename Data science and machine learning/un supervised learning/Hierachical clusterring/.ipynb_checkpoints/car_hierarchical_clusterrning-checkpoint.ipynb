```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import sklearn
```

```python
car_df = pd.read_csv('cluster_mpg.csv')
car_df.head()
```

```python
car_df = car_df.drop('name',axis=1)
```

```python
car_df_dummies = pd.get_dummies(data=car_df,columns=['origin'])
car_df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(car_df_dummies)
scaled_df = pd.DataFrame(scaled_data,columns=car_df_dummies.columns)
```

```python
sns.heatmap(data=scaled_df)
plt.show()
```

```python
sns.clustermap(data=scaled_df,col_cluster=False)
plt.show()
```

```python
sns.clustermap(data=scaled_df,row_cluster=False)
plt.show()
```

```python
sns.clustermap(data=scaled_df)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4)
cluster_labels = model.fit_predict(scaled_df)
car_df['Cluster_labels'] = cluster_labels
car_df
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=np.sqrt(11),linkage='average')

cluster_labels = model.fit_predict(scaled_df)
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage

np.set_printoptions(suppress=True)

linked = linkage(scaled_df,method='ward')
linked
```

```python
plt.figure(figsize=(20,15), dpi=200)
dendrogram(
    linked,
    orientation='top',
    labels=range(0,len(scaled_df)),
    distance_sort='descending'
)
plt.show()
```
