# Exercise 3

# Items data frame

```python
missing_data = items_df_copy.isnull().sum()
percentage_missing_data = missing_data * 100 / items_df_copy.count()
print(percentage_missing_data)

items_df_copy['item_category'] = items_df_copy['item_category'].fillna('unknown')
items_df_copy['item_brand'] = items_df_copy['item_brand'].fillna('unknown')
items_df_copy['stock_quantity'] = items_df_copy['stock_quantity'].fillna(0)
items_df_copy
```

# Exercise 3
# Customers dataframe

```python
missing_data = customers_df_copy.isnull().sum()
percentage_missing_data = missing_data * 100 / customers_df_copy.count()
print(percentage_missing_data)

customers_df_copy['age'] = customers_df_copy['age'].fillna(customers_df_copy['age'].mean())
customers_df_copy['gender']= customers_df_copy['gender'].fillna('unknown')
customers_ids_to_remove = customers_df_copy[customers_df_copy['joining_date'].isna()]
customers_ids_to_remove = customers_ids_to_remove['id']
print(customers_ids_to_remove)
customers_df_copy = customers_df_copy[~customers_df_copy['joining_date'].isna()]
customers_df_copy
orders_df_copy = orders_df_copy[~orders_df_copy['customer_id'].isin(customers_ids_to_remove.values)]
orders_df_copy
```

# Exercise 3
# Order dataframe

```python
missing_data = orders_df_copy.isna().sum()
percentage_missing_data = missing_data * 100 / orders_df_copy.count()
print(percentage_missing_data)

order_ids_to_remove = orders_df_copy[orders_df_copy['customer_id'].isna()]
order_ids_to_remove = order_ids_to_remove['order_id']
order_ids_to_remove.reset_index(name='order_id')

print(f'Order ids to remove: {order_ids_to_remove}')

orders_df_copy = orders_df_copy[~orders_df_copy['customer_id'].isna()]
orders_df_copy['delivey_days'] = orders_df_copy['delivery_days'].fillna(orders_df_copy['delivery_days'].mean())
order_item_df_copy = order_item_df_copy[~order_item_df_copy['order_id'].isin(order_ids_to_remove.values)]
print(orders_df_copy)
order_item_df_copy
```

# Exercise 3

```python
missing_data = order_item_df_copy.isnull().sum()
percentage_missing_data = missing_data * 100 / order_item_df_copy.count()
print(percentage_missing_data)

order_item_df_copy = order_item_df_copy[~order_item_df_copy['item_id'].isna()]
order_item_df_copy
```

# Exercise 4.A

```python
orders_df_copy_total_price = orders_df_copy.copy()
orders_items = orders_df_copy.merge(order_item_df,on='order_id',how='inner').merge(items_df_copy,left_on='item_id',right_on='id',how='inner')
prices = orders_items.groupby('order_id')['item_price'].sum()
prices = prices.reset_index(name='total_price')
orders_df_copy_total_price = prices.merge(orders_df_copy_total_price,on='order_id',how='inner')
print(orders_df_copy_total_price)
```

```python
max_order = round(orders_df_copy_total_price['total_price'].max())
min_order = round(orders_df_copy_total_price['total_price'].min())
mean_order = round(orders_df_copy_total_price['total_price'].mean())
stats = pd.Series({'min order':min_order,'max order':max_order,'avg order':mean_order})
stats = stats.reset_index(name='price')
stats
```
