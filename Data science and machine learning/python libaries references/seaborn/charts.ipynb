```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('csv/dm_office_sales.csv')
df
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='salary',y='sales',hue='training level', palette='dark', s=25, alpha=0.8,marker='x')
plt.title('plot')
plt.show()
```

```python
sns.displot(data=df,x='work experience',kde=True)
plt.show()
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='level of education',hue='training level')
plt.show()
```

```python
sns.barplot(data=df,x='training level',y='sales',palette='Set2',hue='training level')
plt.legend(loc=(1.1,0.5))
plt.show()
```

```python
plt.bar(df['training level'],df['sales'])
plt.show()
```

```python
sns.boxplot(data=df,x='training level',y='sales')
plt.show()
```
