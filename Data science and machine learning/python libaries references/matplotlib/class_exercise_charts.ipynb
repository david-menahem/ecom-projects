```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv/matplotlib_subplot_class_exc.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
usa = df[df['Country'] =='USA']
print(usa)
plt.plot(usa['Year'],usa['GDP per Capita'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
year_2015 = df[df['Year'] == 2015]
```

```python
plt.hist(year_2015['GDP per Capita'])
plt.xlabel('Gdp range')
plt.title('2015')
plt.ylabel('GDP')
plt.show()
```

```python
year_2010 = df[df['Year'] == 2010]
year_2010
plt.bar(year_2010['Country'],year_2010['GDP per Capita'])
plt.title('2010')
plt.ylim(50000,70000)
plt.xlabel('Country')
plt.ylabel('GDP')
plt.show()
```

```python
plt.scatter(usa['Year'],usa['GDP per Capita'])
plt.title('Year/GDP')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```
