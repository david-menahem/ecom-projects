```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
'Manufacturer': ['Toyota', 'Ford', 'Ford', 'Audi', 'BMW', 'Toyota', 'Audi', 'BMW',
'Nissan', 'Nissan'],
'Type': ['Sedan', 'Truck', 'SUV', 'Sedan', 'Sedan', 'Truck', 'SUV', 'Truck',
'SUV', 'Sedan'],
'Fuel_Efficiency': [34, 22, 24, 27, 28, 24, 26, 21, 23, 35],
'Horsepower': [180, 320, 260, 220, 240, 290, 210, 310, 250, 200],
'Price': [30000, 40000, 35000, 45000, 50000, 32000, 47000, 48000, 33000,
31000]
}
df = pd.DataFrame(data)
print(df)
```

# Exercise 2

```python
df.sort_values(by='Horsepower',inplace=True)
plt.plot(df['Horsepower'],df['Price'],color='red',ls='--',marker='.')
plt.ylabel('Price')
plt.xlabel('Horse Power')
plt.title('Price vs Horse Power')
# plt.xlim(30000,50000)
price_horsepower_line = plt.show()
```

# Exercise 3

```python
by_car_type = df.groupby('Type').mean('Type')
types = by_car_type.index.values
plt.bar(types,by_car_type['Price'],color=['red','blue','orange'])
plt.xlabel('Car type')
plt.ylabel('Avg price')
plt.ylim(0,50000)
plt.show()
```

# Exercise 4

```python
types = by_car_type.index
colors = ['red','orange','blue']
plt.bar(types.values,by_car_type['Horsepower'],color=colors,label=types.values)
plt.xlabel('Car type')
plt.ylabel('Avg horsepower')
plt.title('Average horsepower by car type')
plt.legend()
plt.show()
```

# Exercise 5

```python
plt.hist(df['Fuel_Efficiency'],bins=5)
plt.xlabel('Fuel Efficiency Range')
plt.ylabel('Count')
plt.title('Distrubution of fuel efficiency')
plt.show()
```

# Exercise 6

```python
plt.scatter(df['Horsepower'],df['Price'])
plt.xlabel('Price')
plt.ylabel('Horsepower')
plt.title('There is no corrilation between horse power and its price')
plt.show()
```

# Exercise 7

```python
fig,axes_arr = plt.subplots(nrows=2,ncols=2)
plt.subplots_adjust(wspace=0.5,hspace=0.5)
axes_arr[1][1].plot(df['Horsepower'],df['Price'],color='red',ls='--',marker='.')
axes_arr[1][1].set_ylabel('Price')
axes_arr[1][1].set_xlabel('Horse Power')
axes_arr[1][1].set_title('Price vs Horse Power')

axes_arr[1][0].bar(types.values,by_car_type['Horsepower'],color=colors,label=types.values)
axes_arr[1][0].set_xlabel('Car type')
axes_arr[1][0].set_ylabel('Avg horsepower')
axes_arr[1][0].set_ylim(0,500)
axes_arr[1][0].set_title('Average horsepower by car type')
axes_arr[1][0].legend(fontsize='small')

axes_arr[0][1].hist(df['Fuel_Efficiency'],bins=5)
axes_arr[0][1].set_xlabel('Fuel Efficiency Range')
axes_arr[0][1].set_ylabel('Count')
axes_arr[0][1].set_title('Distrubution of fuel efficiency')

axes_arr[0][0].scatter(df['Horsepower'],df['Price'])
axes_arr[0][0].set_xlabel('Price')
axes_arr[0][0].set_ylabel('Horsepower')
axes_arr[0][0].set_title('No corrilation with horse power vs price')

fig.savefig('exported_images/cars.jpeg',bbox_inches='tight')
```
