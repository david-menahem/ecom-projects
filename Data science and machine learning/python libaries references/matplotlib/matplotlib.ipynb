MatPlotLib is a 2D plotting library
line plots, graph, histogram, scatter plots..

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2 * x
a = np.arange(0,20)
b = a ** 5
plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X data')
plt.ylabel('Y label')
plt.xlim(0,6)
plt.ylim(0,12)
plt.savefig('exported_images/chart.jpeg')
plt.show()
```

```python
fig = plt.figure(figsize=(10,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
axes.set_title('power Chart')
axes.set_xlabel('A')
plt.ylabel('Y')
second_axes = fig.add_axes([0.2,0.5,0.25,0.25])
second_axes.set_title('zoomed in')
second_axes.plot(a,b)
second_axes.set_xlim(10,12)
fig.savefig('exported_images/power.jpg',bbox_inches='tight')
plt.show()
```

```python
x1 = np.arange(10)
y1 = x1*2
x2 = np.arange(20)
y2 = np.sin(x2)
```

Subplots

```python
fig,axes_arr = plt.subplots(nrows=2,ncols=2)
# plt.subplots_adjust(left=1,right=2,top=2,bottom=1)
plt.subplots_adjust(wspace=0.4,hspace=0.4)
#fig.tight_layout()
fig.suptitle('plots',fontsize=15)
axes_arr[0][0].plot(x1,y1)
axes_arr[0][0].set_title('plot 1')

axes_arr[0][1].plot(x2,y2)
axes_arr[0][1].set_title('plot 2')

axes_arr[1][0].plot(x1,y1)
axes_arr[1][0].set_title('plot 3')

axes_arr[1][1].plot(x2,y2)
axes_arr[1][1].set_title('plot 4')
```

2 graphs in one plot

```python
x1 = np.arange(10)
y1 = x1*2
x2 = np.arange(20)
y2 = np.sin(x2)

fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
axis.plot(x1,y1,label='Linear',color='red',lw=4,ls='--',marker='o',markersize=15)
axis.plot(x2,y2,label='sin',color='purple',lw=5,ls=':',marker='+',markersize=10)
axis.set_title('2 graphs')
axis.legend(loc=(0.85,0.9))

plt.show()
```
