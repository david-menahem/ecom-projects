```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.linspace(0.0,5.0,51,dtype=float)
y_1 = np.sin(arr)
y_2 = np.cos(arr)
```

```python
fig = plt.figure(figsize=(10,10))
axes_1 = fig.add_axes([0,0,0.4,0.5])
axes_1.set_title('sinus')
axes_1.set_xlabel('angle')
axes_1.set_ylabel('sin')

axes_2 = fig.add_axes([0.6,0,0.4,0.5])
axes_2.set_title('cosinus')
axes_2.set_xlabel('angle')
axes_2.set_ylabel('cos')

axes_1.plot(arr,y_1)
axes_2.plot(arr,y_2)
plt.show()
fig.savefig('sinus_cosinus.jpg',bbox_inches='tight')
```
