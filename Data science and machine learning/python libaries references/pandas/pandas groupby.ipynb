```python
import numpy as np
import pandas as pd
```

```python
df_mpg = pd.read_csv('csv/mpg.csv')
print(df_mpg)
```

```python
grouped_df = df_mpg.groupby('model_year').mean('model_year')
print(grouped_df)
print(grouped_df.loc[70])
```

```python
count_grouped_model = df_mpg['model_year'].value_counts().sort_index(ascending=False)
print(count_grouped_model)
```

```python
multi_index_df = df_mpg.groupby(['model_year','cylinders']).mean(['model_year','cylinders'])
print(multi_index_df)
print(multi_index_df.index)
```

```python
print(multi_index_df.loc[(78,4)])
print(multi_index_df.loc[72])
print(multi_index_df.xs(key=8,level=1))
print("....")
print(multi_index_df.xs(key=8,level='cylinders'))
```

```python
print(df_mpg[df_mpg['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean(['model_year','cylinders']))
print(multi_index_df.loc[[70,71]])
```

```python
multi_index_df.swaplevel()
```

```python
multi_index_df.sort_index(level=[0,1],ascending=[True,False])
multi_index_df.swaplevel().sort_index()
print(multi_index_df.swaplevel().loc[[6,8]])
```

```python
agg_function_dict = {'mpg':['max','mean'],'weight':['mean','std']}
df_mpg_stats = df_mpg.agg(agg_function_dict)
print(df_mpg_stats)
```

```python
df_1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],'B': ['B0','B1','B2','B3'],'C':['C0','C1','C2','C3']})
df_2 = pd.DataFrame({'A': ['A4','A5','A6','A7'],'B': ['B4','B5','B6','B7'],'C':['C4','C5','C6','C7']})

print(df_1)
print()
print(df_2)

concat_row_df = pd.concat([df_1,df_2]).reset_index()
print()
print(concat_row_df)

print()
df_3 = pd.DataFrame({'A': ['A0','A1','A2','A3'],'B': ['B0','B1','B2','B3'],'C':['C0','C1','C2','C3']})
df_4 = pd.DataFrame({'D': ['D4','D5','D6','D7'],'E': ['E4','E5','E6','E7'],'F':['F4','F5','F6','F7']})
print(df_3)
print()
print(df_4)
print()
concat_col = pd.concat([df_3,df_4],axis=1)
print(concat_col)
```

```python
df_5 = pd.DataFrame({'A': ['A0','A1','A2','A3'],'B': ['B0','B1','B2','B3'],'C':['C0','C1','C2','C3']})
df_5.set_index('A',inplace=True)
df_6 = pd.DataFrame({'D': ['D4','D5','D6','D7'],'E': ['E4','E5','E6','E7'],'F':['F4','F5','F6','F7']})
merged_df = pd.merge(df_5,df_6,left_index=True,right_on='D',how='outer')
print(merged_df)
```
