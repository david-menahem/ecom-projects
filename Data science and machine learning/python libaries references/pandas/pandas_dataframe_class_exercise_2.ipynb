```python
def order_summary(order,store_series):
    order_sum = 0
    
    order_summary = {'total_pay': 0,
                    'available_stores':{}}
    available_stores = []
    store_count = 0
    stores = store_series.keys()
    print(stores)
    for item in order:
        found = False
        bought = False
        for store in store_series:
            if item in store:
                if not bought:
                    order_sum += store[item]
                    bought = True
                available_stores.append(stores[store_count])
                found = True
            store_count += 1
        if not found:
            available_stores.append("NaN")
            order_summary['available_stores'][item] = available_stores
        store_count = 0
        bought = False
    order_summary['total_pay'] = order_sum

    return order_summary

order_result = order_summary(order_arr,store_series)
print(f"The order price is: {order_result['total_pay'].round(2)}\n The places that the items were located are: \n{order_result['available_stores']}")
```

```python
tel_aviv = pd.Series({'apple': 10, 'banana': 12, 'orange': 8, 'kiwi': 20, 'watermelon': 13 })
haifa = pd.Series({'apple': 10, 'carrot': 11, 'orange': 8, 'eggplant': 25, 'watermelon': 13 })
ramat_gan = pd.Series({'lettuce': 4, 'ginger': 33, 'orange': 8, 'kiwi': 20, 'garlic': 3 })
stores = {'Tel Aviv': tel_aviv, 'Haifa': haifa, 'Ramat Gan': ramat_gan}

def order_purchase(product_list):
    total_pay = 0
    available_stores = {}
    
    for product in product_list:
        pay_added = False
        product_available_stores = []

        for store_name, store_data in stores.items():
            if product in store_data.keys():
                product_available_stores.append(store_name)
                if not pay_added:
                    total_pay += store_data[product]
                    pay_added = True

        if (len(product_available_stores) > 0):
            available_stores[product] = product_available_stores
        else:
            available_stores[product] = None

    return {'total_pay': total_pay, 'available_stores': available_stores}

print(order_purchase(['apple','carrot','orange','ice cream']))
```
