```python
import numpy as np
import pandas as pd
```

```python
tips_df = pd.read_csv('csv/tips.csv')
print(tips_df)
```

```python
def calculate_discount(total_bill,tip,size):
    if size >= 4:
        discount = (total_bill + tip) / 10
        return round(discount,2)
    else: return 0.0
tips_df['discount'] = np.vectorize(calculate_discount)(tips_df['total_bill'],tips_df['tip'],tips_df['size'])
print(tips_df)
```

```python
def convert_day(day):
    days = {'Sun':1,
            'Mon':2,
            'Tues':3,
            'Wed':4,
            'Thur':5,
            'Fri':6,
            'Sat':7}
    if day in days:
        return days[day]
    

tips_df['day in number'] = tips_df['day'].apply(convert_day)
tips_df.sort_values(by=['day in number','total_bill'],ascending=[True,False]).head(20)
```

```python
def adjust_bill_dinner(total_bill,time):
    if time == "Dinner":
        return round(total_bill * 1.05,2)
    else: return total_bill
tips_df_copy = tips_df.copy()
tips_df_copy.loc[tips_df['time'] == 'Dinner','total_bill'] = tips_df_copy.apply(lambda row : adjust_bill_dinner(row['total_bill'],row['time']),axis=1)
print(tips_df_copy)  
```

```python
data = pd.DataFrame({'A':[1,2,3,np.nan,4],
                    'B':[6,7,8,np.nan,np.nan],
                    'C':[11,13,13,np.nan,15]})
print(data)
print(data.isnull())
data['A'] = data['A'].fillna(data['A'].mean())
data['B'] = data['B'].fillna(data['B'].median())
data['C'] = data['C'].fillna(data['C'].mode()[0])
print(data)
```

```python
grouped_tips_df = tips_df.groupby('day').agg({'total_bill':'mean','tip':'mean'}).reset_index()
print(grouped_tips_df)
merged_tips_df = pd.merge(tips_df,grouped_tips_df,on='day',suffixes=['','_avg'])
print(merged_tips_df)
```

```python
def calculate_mean(col):
    if isinstance(col.iloc[0],(int,float)):
        return col.mean()
    else:
        return "not digit"
def replace_based_on_condition(mean,value):
    if value >= mean:
        return "High"
    else:
        return "Low"
tips_df_copy = tips_df.copy()

for col in tips_df_copy:
    mean = calculate_mean(tips_df_copy[col])
    if mean != "not digit":
        tips_df_copy[col] = np.vectorize(replace_based_on_condition)(mean,tips_df_copy[col])
print(tips_df_copy)
```

```python
from datetime import datetime, timedelta

# Generate synthetic time series data
start_date = datetime(2022, 1, 1)
end_date = start_date + timedelta(days=365)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

temperature_data = np.random.uniform(20, 30, size=len(date_range))

time_series_df = pd.DataFrame({'Date': date_range, 'Temperature': temperature_data})
print(time_series_df)

grouped_df = time_series_df.resample('W-Mon', on='Date').mean()
print(grouped_df)
```
