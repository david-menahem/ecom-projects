```python
import numpy as np
import pandas as pd
```

```python
tips_df = pd.read_csv('csv/tips.csv')
```

```python
print(tips_df)
```

```python
def last_four(cc):
    return int(str(cc)[-4:])
print(last_four(5296068606052842))
```

```python
tips_df['cc last 4 digits'] = tips_df['CC Number'].apply(last_four)
print(tips_df)
```

```python
def is_genouras(total_bill,tip):
    if tip * 100 / total_bill > 15:
        return "Genouras"
    else:
        return "Not genouras"
    
tips_df['is_genouras'] = np.vectorize(is_genouras)(tips_df['total_bill'],tips_df['tip'])
print(tips_df)
```

```python
print(tips_df.sort_values(by=['tip','total_bill'],ascending=[True,False]))
```

```python
print(tips_df[['is_genouras','sex','smoker']].value_counts())
```

```python
print(tips_df['smoker'].unique())
print(tips_df['smoker'].nunique())
```

```python
tips_df_copy = tips_df.copy()
tips_df_copy['sex'] = tips_df_copy['sex'].replace(['Female','Male'],['F','M'])
print(tips_df_copy)
```

```python
tips_df_copy = tips_df.copy()
replace_dict = {'Female': 'F','Male': 'M'}
tips_df_copy['sex'] = tips_df_copy['sex'].map(replace_dict)
print(tips_df_copy)
```
