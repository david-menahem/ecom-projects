```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")

titanic
```

```python
columns = ['age', 'fare', 'survived']
df = titanic[columns]

df = df.dropna()

df
```

```python
sns.countplot(data=df, x='survived')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

X = df[['age', 'fare']]
y = df['survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

logistic_model = LogisticRegression()
logistic_model.fit(scaled_X_train, y_train)
```

```python
logistic_model.coef_
```

```python
y_pred = logistic_model.predict(scaled_X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy: {accuracy}')

percision = precision_score(y_test, y_pred)
print(f'percision: {percision}')

recall = recall_score(y_test, y_pred)
print(f'recall: {recall}')

# f1_scoring = f1_score(y_test, y_pred)
# print(f'f1_score: {f1_scoring}')


print(confusion_matrix(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cmap = plt.cm.viridis

sns.heatmap(cm, annot=True, fmt="d", cmap=cmap)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
