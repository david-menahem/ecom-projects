```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.countplot(data=df, x='species')
plt.show()
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV

X = df.drop('species', axis=1)
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)


scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)


logistic_model = LogisticRegressionCV(cv=5, max_iter=1000, multi_class='multinomial', 
                                      solver='saga', penalty='l2', Cs=np.logspace(0,4,10)
                                     )

logistic_model.fit(scaled_X_train, y_train)
```

```python
logistic_model.coef_
```

```python
y_pred = logistic_model.predict(scaled_X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
cmap = plt.cm.viridis

sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap=cmap)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
