```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins_df_origianl = pd.read_csv('penguins_size.csv')
penguins_df = penguins_df_origianl.copy()
penguins_df.head()
```

```python
penguins_df = penguins_df.dropna()
print(len(penguins_df))
```

```python
penguins_df_copy = penguins_df.copy()
for column in penguins_df_copy.columns:
    penguins_df_copy[column] = pd.Series(penguins_df_copy[column].unique())
    
print(penguins_df_copy.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix


X = penguins_df.drop('species', axis=1)
X = pd.get_dummies(X,columns=['island','sex'])
X = X.drop('sex_FEMALE',axis=1)
y = penguins_df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tree_model = DecisionTreeClassifier()

tree_model.fit(X_train, y_train)

pred = tree_model.predict(X_test)
pred
```

```python
print(confusion_matrix(y_test, pred))
print()
print(accuracy_score(y_test, pred))
print()
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8), dpi=200)
plot_tree(tree_model, feature_names=X.columns.tolist(), filled=True)
plt.show()
```
