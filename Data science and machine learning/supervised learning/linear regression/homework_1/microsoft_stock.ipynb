# Data preperation

```python
microsoft_df_copy['Date'] = pd.to_datetime(microsoft_df_copy['Date'])
microsoft_df_copy_date_index = microsoft_df_copy.set_index('Date')
```

```python
X = microsoft_df_copy.drop(['Date','Close','Adj Close'],axis=1)
y = microsoft_df_copy.drop(['Date','Open','High','Low','Volume','Adj Close'],axis=1)
```

# Linear regression

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
test_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)
print(f'MAE: {MAE}\nMSE: {MSE}\nRMSE: {RMSE}')
```

```python
mae_percentage = MSE * 100 / microsoft_df_copy_date_index['Close'].mean()
rmse_percnetage = RMSE * 100 / microsoft_df_copy_date_index['Close'].mean()
print(f'Error percentages:\nMAE: {mae_percentage}\nRSME: {rmse_percnetage}')
```

# All though the percentages are low considering the low percentage a stock changes in one day it is not acceptable

```python
test_residuals = y_test - test_predictions
sns.scatterplot(x=y_test['Close'],y=test_residuals['Close'])
plt.axhline(y=0, color='r', ls='--')
plt.show()
```

# The plot shows that the errors are very small but when the value is high the error get much bigger so 
# I am still not convinced this model is the best solution

# Deployment

```python
from joblib import dump,load
final_model = LinearRegression()
final_model.fit(X,y)
dump(final_model,'models/microsoft_stock_predictions')
```
