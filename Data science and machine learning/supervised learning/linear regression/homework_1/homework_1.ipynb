```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tempratures = pd.Series([10,20,30,40,50,60,70,80,90])
hours = pd.Series([420,365,285,220,176,117,69,34,5])
df = pd.DataFrame({'hours': hours, 'temprature': tempratures})
df_copy = df.copy()
df
```

```python
plt.figure(figsize = (10,10))
sns.regplot(data = df_copy, x='temprature', y='hours')
plt.show()
```

# There is a clear linear correlation between the tempretures and the hours of using the machine

```python
my_betas = [-5.31,457.7]
np_betas = np.polyfit(df_copy['temprature'],df_copy['hours'], deg=1)
tempratures = np.linspace(0,100,20)
hours = tempratures * my_betas[0] + my_betas[1]
plt.plot(tempratures,hours)
plt.ylabel('hours')
plt.xlabel('tempratures')
plt.title('The linear regression of the hours dependent of the tempreture')
plt.show()
print(np_betas)
```

# Question G

```python
tv_advertising = np.linspace(0,300,25)
sales = betas_tv[3] + betas_tv[2] * tv_advertising +  betas_tv[1] * tv_advertising**2 + betas_tv[0] * tv_advertising**3
plt.plot(tv_advertising,sales)
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('The linear regression representing Sales depending on TV advertising')
plt.show()
```

```python
radio_advertising = np.linspace(0,100,25)
sales = betas_radio[3] + betas_radio[2] * radio_advertising +  betas_radio[1] * radio_advertising**2 + betas_radio[0] * radio_advertising**3
plt.plot(radio_advertising,sales)
plt.xlabel('radio')
plt.ylabel('Sales')
plt.title('The linear regression representing Sales depending on radio advertising')
plt.show()
```

```python
newspaper_advertising = np.linspace(0,50,25)
sales = betas_newspaper[3] + betas_newspaper[2] * newspaper_advertising +  betas_newspaper[1] * newspaper_advertising**2 + betas_newspaper[0] * newspaper_advertising**3
plt.plot(newspaper_advertising,sales)
plt.xlabel('newspaper')
plt.ylabel('Sales')
plt.title('The linear regression representing Sales depending on newspaper advertising')
plt.show()
```
