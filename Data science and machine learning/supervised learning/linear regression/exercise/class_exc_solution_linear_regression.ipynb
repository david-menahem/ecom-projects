```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hours = np.array([2, 3, 4, 5, 6, 1.5, 5, 7, 8, 10])
money_earned = np.array([50, 70, 90, 100, 110, 40, 110, 130, 145, 180])

hours_reshape = hours.reshape(-1, 1)
money_earned_reshape = money_earned.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(hours_reshape, money_earned_reshape, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, y_predictions)
mse = mean_squared_error(y_test, y_predictions)
rmse = np.sqrt(mse)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
y_predictions
```

```python
mae_precent = (mae * 100 / money_earned.mean())
rmse_precent = (rmse * 100 / money_earned.mean())

print(f'Precent MAE: {mae_precent}')
print(f'Precent RMSE: {rmse_precent}')
```

```python
residuals = y_test - y_predictions
plt.scatter(y_test, residuals)
plt.axhline(y=0, color='r', ls='--')
plt.xlabel("Actual Money Earned (Thousands of $)")
plt.ylabel("Residuals")
plt.title("Residuals vs Actual Money Earned")
plt.show()
```
