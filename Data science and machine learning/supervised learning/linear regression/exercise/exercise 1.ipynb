```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hours = pd.Series([2,3,4,5,6])
earnings = pd.Series([50,70,90,100,110])
df = pd.DataFrame({"hours":hours,"earnings":earnings})
df
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data =df, x='hours', y='earnings')
plt.xlabel('Calling hours')
plt.ylabel('Earnings')
plt.show()
```
