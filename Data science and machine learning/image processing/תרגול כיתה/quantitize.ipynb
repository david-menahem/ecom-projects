```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.image as mpimg
image_as_array = mpimg.imread('WhatsApp Image 2024-08-04 at 20.46.22.jpeg')
image_as_array
```

```python
(h,w,c) = image_as_array.shape
flatten_image_array = image_as_array.reshape(h*w,c)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=50)
labels = model.fit_predict(flatten_image_array)
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
new_flatten_array = rgb_codes[labels]
new_flatten_array
```

```python
new_image_array = new_flatten_array.reshape(h,w,c)
```

```python
plt.imshow(image_as_array)
plt.show()
```

```python
plt.imshow(new_image_array)
plt.show()
```
