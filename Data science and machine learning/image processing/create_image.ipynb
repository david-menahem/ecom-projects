```python
from matplotlib import image as mpimg
image_as_array = mpimg.imread('WhatsApp Image 2024-08-04 at 19.29.37.jpeg')
image_as_array
```

```python
(h,w,c) = image_as_array.shape
```

```python
reshaped_image_array = image_as_array.reshape(h*w,c)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=20)
labels = model.fit_predict(reshaped_image_array)
labels
```

```python
from collections import Counter
counter = Counter(labels)
counter
```

```python
colors = model.cluster_centers_
```

```python
rgb_codes = colors.round(0).astype(int)
rgb_codes
```

```python
flatten_new_array = rgb_codes[labels]
flatten_new_array
```

```python
new_image_array = flatten_new_array.reshape(h,w,c)
new_image_array
```
